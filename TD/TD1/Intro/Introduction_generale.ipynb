## Apprentissage supervisé, apprentissage non supervisé

Les problèmes qui relèvent de l'apprentissage automatique peuvent être divisés en deux grandes catégories suivant que l'on guide ou non l'apprentissage en lui fournissant sur une partie des données la réponse attendue du problème. On distingue donc l'apprentissage supervisé (supervised learning) pour lequel on dispose des données complètes ou couples d'information (paramètres d'entrée, classe ou paramètres de sortie) de l'apprentissage non supervisé pour lequel seuls les paramètres d'entrée (extraits des données brutes) sont disponibles. L'apprentissage supervisé recouvre (entre autres) les problèmes suivants:

- La Prédiction : le paramètre de sortie est de type quantitatif 
- La Discrimination (Classification en anglais): le paramètre de sortie est de type catégoriel 

Un exemple d'apprentissage supervisé est celui d'un Filtre de SPAMS : son entrée brute est un email et la sortie désirée est la décision SPAM ou NON-SPAM (c'est de la discrimination).
L'ensemble des paramètres d'entrée pourra être un vecteur booléen $X=(x_1,...,x_i,...,x_N)$ où $x_i = 1$ (true) si le i-ème mot du dictionnaire apparait dans le mail et 0 (false) sinon, N étant le nombre de mots du dictionnaire considéré (quelques centaines ou milliers). Un système d'apprentissage automatique pour cette application aura comme base de données d'apprentissage un grand nombre de couples $(X_k,y_k)$ où $X_k$ est le vecteur booléen associé à un mail et $y_k$ la sortie (SPAM ou NON-SPAM) correspondante.

L'apprentissage non supervisé (unsupervised learning) est le plus souvent lié à un problème de Classification (Clustering en anglais). Il s'agit alors pour la machine de dégager elle même les ressemblances ou dissemblances entre données et d'en déduire des groupes ou classes ayant une certaine cohérence. Nous mettrons en oeuvre ce type d'apprentissage lors de la première application qui consistera à regrouper par thèmes des messages, sans que ces thèmes aient été préalablement définis. L'apprentissage non supervisé permet aussi de réduire la dimension des données en prenant en compte la corrélation entre les paramètres d'entrée. 

## Composantes d'un système d'apprentissage automatique




Il y a trois composantes principales dans un système d'apprentissage automatique:

-  L'extraction de paramètres (feature engineering en anglais). Il s'agit d'extraire des données brutes telles que celles que nous avons citées plus haut des valeurs numériques appelées **paramètres ou attributs ou variables** regroupées le plus souvent en vecteurs, chaque donnée étant associée à un **vecteur de paramètres** qui sera l'entrée du système d'apprentissage automatique.  C'est une étape clé qui demande de l'intuition, de la créativité et du savoir faire qui vient de l'expérience. Le plus souvent (ce sera toujours vrai sur les données web qui nous occuperons dans ce cours) les données brutes sont inexploitables, il faut en extraire des paramètres (quantitatifs généralement) qui soient a priori pertinents pour la tâche que l'on souhaite accomplir, c'est à dire qui "aident" l'algorithme à prédire, classer... Un domaine de recherche en Machine Learning consiste justement à chercher des méthodes permettant d'automatiser ou d'optimiser ce processus d'extraction de paramètres.
-  Le choix de la technique utilisée. Pour une même tache (par exemple la discrimination) on peut mettre en oeuvre plusieurs techniques (la méthode des plus proches voisins, l'approche bayésienne, la régression logistique) et il faut éventuellement tester plusieurs techniques pour trouver celle qui est la plus adaptée aux données (et à leur dimension). 
-  L'évaluation de l'apprentissage. Les performances peuvent être mesurées par différents critères tels que le taux d'erreur, l'erreur quadratique moyenne, la probabilité d'erreur, la vraisemblance. Il est plus difficile de mesurer des performances pour l'apprentissage non supervisé qui sert souvent d'étape préalable à un apprentissage supervisé.





## Evaluation de l'apprentissage, under/overfitting, biais et variance, performances


Pour toutes les techniques faisant appel à un apprentissage supervisé (prédiction et discrimination principalement) le but ultime est leur capacité de généralisation au delà des exemples fournis lors de l'apprentissage. C'est à dire que la prédiction ou la discrimination doit être fiable pour des données (paramètres d'entrée) qui n'ont pas été rencontrés lors de la phase d'apprentissage. 
Exemple:
Si l'on considère l'exemple mentionné plus haut, relatif au filtrage des spams, le nombre de données possibles est de 2^N, N étant le nombre de mots du dictionnaire (quelques centaines ou milliers) et il est évidemment non concevable de fournir un nombre d'exemples du même ordre de grandeur, typiquement on pourra fournir de l'ordre de 10^4 à 10^6 exemples. Le problème est alors que les performances peuvent être excellentes sur les données utilisées pour l'apprentissage et s'effondrer pour de nouvelles données test. 
C'est ce qu'on appelle **l'overfitting** (très bonne mémorisation des exemples mais mauvaise généralisation).




###  Illustration du problème d'overfitting ou underfitting

On dispose de données (X,Y) telles que Y=f(X)+ bruit (f est normalement inconnue, mais pas ici, pour pouvoir visualiser les performances ). On cherche à "estimer" la fonction f, à partir des couples (X,Y) disponibles de façon à pouvoir prédire Y pour de nouvelles valeurs de X. Pour ce faire on imagine des formes de complexité croissante pour l'estimée de f (polynômes de degré 1,4 ou 15).

```python
from numpy.polynomial import polynomial as P
from numpy.polynomial.polynomial import polyval
n_samples = 30
degrees = [1, 4, 15]
plt.figure(figsize=(14, 5))
true_fun = lambda X: np.cos(1.5 * np.pi * X)

N_iter=10
for j in range(N_iter):
    X = np.sort(np.random.rand(n_samples))
    y = true_fun(X) + np.random.randn(n_samples) * 0.1
    for i in range(len(degrees)):
        ax = plt.subplot(1, len(degrees), i + 1)
        plt.setp(ax, xticks=(), yticks=())
        c = P.polyfit(X,y,degrees[i])
        X_test = np.linspace(0, 1, 100)
        plt.plot(X_test, polyval(X_test,c), label="Model")
        plt.plot(X_test, true_fun(X_test), label="True function")
        plt.scatter(X, y, label="Samples",s=10,alpha=0.5,c='r')
        plt.xlabel("x")
        plt.ylabel("y")
        plt.xlim((0, 1))
        plt.ylim((-2, 2))
        if N_iter==1:
            plt.legend(loc="best")
```

- Si la complexité est faible (polynôme de degré 1) il y a underfitting, la prédiction est mauvaise (le biais est important) mais f estimée est relativement stable (faible variance).

- Si la complexité est grande (polynôme de degré 15) il y a overfitting , la prédiction "colle" aux données Y bruitées (faible biais) mais ne permet pas de généraliser (d'être correcte pour des données X non rencontrées pendant l'apprentissage). Si on change de jeu de données (suivant toujours Y=f(X)+ bruit) f estimée varie fortement (forte variance).

Cet exemple simple est riche d'enseignements. Observons ce qui se passe si nous disposons de 300 couples (X,Y) au lieu de 30 pour l'apprentissage.


Pour contrôler l'écueil du sur-apprentissage (modèle trop complexe, en tous cas en relatif par rapport au nombre de données disponibles) il faut toujours garder une portion des données disponible pour le test (ou la validation) afin de vérifier les capacités de généralisation de l'apprentissage.

###  EQM, Biais, Variance.

Nous avons cité plusieurs critères de performances. L'un des plus souvent utilisés, lorsque la sortie est de type quantitatif, est **l'erreur quadratique moyenne** qui comporte deux termes liés à la variance et au biais respectivement. Le programme ci dessous illustre la signification des termes de biais et variance sur l'exemple d'un tir de fléchettes sur une cible: 

```python
fig=plt.figure(figsize=(14, 14))
labels=['faible biais - faible variance','faible biais - forte variance','fort biais - faible variance','fort biais - forte variance']
i=1
for m in [0,-5]:
    for s in [1,4]:
        
        x=np.random.normal(m,s,(20,1))
        y=np.random.normal(m,s,(20,1))
        z=x+1j*y
        ax = fig.add_subplot(2, 2, i, projection='polar')
        #ax=plt.subplot(22i,polar=True)
        rond=np.linspace(0,2*np.pi,100)
        
        ax.plot(rond,10*np.ones(np.shape(rond)))
        ax.scatter(np.angle(z),np.abs(z), color='blue', lw=1, label=labels[i-1])
        ax.scatter(np.angle(np.mean(z)),np.abs(np.mean(z)),color='red', lw=2)
        ax.legend()
        i=i+1          
```

L'équation sous jacente est:  $ E((y-\hat{y})^2)= E((y-E(\hat{y}))^2)+E((E(\hat{y})-\hat{y})^2)$

- Le terme à gauche du signe égalité est l'Erreur Quadratique Moyenne
- Le premier terme à droite s'écrit encore $(y-E(\hat{y}))^2$ car ici y (la cible) est supposée déterminée (non aléatoire). C'est le biais au carré, le biais mesurant l'écart entre la vraie valeur et la moyenne de son estimée.
- Le second terme à droite est la variance qui mesure l'écart moyen carré de l'estimée à sa moyenne.

Il y a un biais important lorsque le modèle sous jacent de l'algorithme utilisé est trop simple (under-fitting). Il y a une variance importante pour des modèles complexes qui risquent d'être sur-appris (over_fitting) sur les données d'apprentissage et sans capacité de généralisation. 

Faire le lien avec l'illustration under/overfitting précédente en prenant par exemple N_iter=10 (10 estimées).

Le but ultime d'une technique serait bien sûr d'obtenir un faible biais et une faible variance. En réalité il y a toujours un compromis à faire entre ces deux valeurs. 

### Taux d'erreur - Matrice de Confusion

Le critère utilisé le plus souvent pour mesurer les performances lorsque la sortie est de type qualitatif est le **taux d'erreur**. Ce critère peut être subdivisé en plusieurs autres suivant le problème de départ. Imaginons par exemple un diagnostic maligne/bénigne pour une tumeur . Les fausses décisions maligne à la place de bénigne ou bénigne à la place de maligne n'ont pas les mêmes conséquences, ont des "coûts" a priori différents.
Plus généralement lorsque la décision est binaire, de type positif/négatif on mesure le nombre d'occurences de chacun des quatre cas possibles que l'on rassemble dans la **matrice de confusion** suivante:

 $\begin{bmatrix}Predit|Vrai & 0 &1\\0&TN & FN \\1& FP & TP \end{bmatrix} $

Il peut être, suivant les applications, plus important de minimiser le taux de Fausse Alarme FP (False Positive) ou de minimiser le taux de Non Detection FN (False Negative) que de minimiser la somme de ces deux termes d'erreurs.
Ces quatre cas permettent de calculer les critères suivants, fréquemment utilisés pour mesurer les performances:

- L'exactitude (accuracy) : $ (TP+TN)/ (TP+TN+FP+FN) $
- La précision (precision) : $ TP/(TP+FP) $
- La sensibilité (ou taux de vrais positifs) : $ TP/(TP+FN) $
- La spécificité (ou taux de vrais négatifs): $TN/(TN+FP) $ 

Un classificateur avec un faible pourcentage d'erreur (une très bonne exactitude) peut s'avérer tout à fait défaillant pour détecter la plupart des P (mauvaise sensibilité) si la plupart des cas sont N.

## Premier Exemple: Discrimination d'Iris par seuillage


Il s'agit de distinguer trois espèces d'iris (Virginica, Setosa, Versicolor) à partir de quatre caractéristiques (ou paramètres) sur ces fleurs (la longueur et la largeur du sépale, la longueur et la largeur du pétale) [sépale/pétale](http://upload.wikimedia.org/wikipedia/commons/thumb/7/78/Petal-sepal.jpg/300px-Petal-sepal.jpg).
C'est un exemple classique que l'on retrouve dans la plupart des livres ou articles en Machine Learning: [Iris]( http://en.wikipedia.org/wiki/Iris_flower_data_set). La base des Iris est en particulier disponible comme exemple dans la librairie de Scikit-learn.

Dans ce premier cas il n'y a donc pas de problème d'extraction de paramètres (feature engineering). Il n'y a pas non plus de vrai problème de classification car on connait déja l'espèce de chacune des 150 fleurs de la base. On peut cependant à partir de tout ou partie de cette base établir les règles permettant de classer par espèce un iris du reste de la base si on ne l'a pas utilisée entièrement ou un iris ne faisant pas partie de la base. C'est donc un cas d'apprentissage supervisé.

### Visualisation de la base

En raison de la petite taille de cette base de données il est possible et utile de représenter les données dans chacun des 6 plans représentant toutes les combinaisons de paramètres possibles.

```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
feature_names = iris.feature_names
print(feature_names)
print(target_names)
print('taille de la base :{}'.format(X.shape))
```

```python
def plot_croises(X,y,target_names,feature_names):
    nb_class=len(np.unique(y))
    nb_feat=len(feature_names)
    if nb_feat%2:
        nrows=int((nb_feat-1)/2)
        ncols=nb_feat
    else:
        nrows=int(nb_feat/2)
        ncols=nb_feat-1
    plt.figure(figsize=(14, 10))
    col=['b','r','g','c','m','y','k']
    for t in range(nb_class):
        k=1
        for i in range(nb_feat-1):
            for j in range(i+1,nb_feat):
                plt.subplot(nrows,ncols,k)
                plt.scatter(X[y == t,i],X[y == t,j],color=col[t],lw=2,marker='o',label=target_names[t])
                plt.ylabel(feature_names[j])
                plt.xlabel(feature_names[i])
                k=k+1
    plt.legend(bbox_to_anchor=(2,2))  
    plt.tight_layout()  
    
```

```python
plot_croises(X,y,target_names,feature_names)
```

### Premières observations et classification élémentaire par seuillage

Les Iris Setosa (triangles rouge) apparaissent clairement séparés des deux autres espèces Versicolor (cercles verts) et Virginica (croix bleues) qui semblent un peu moins aisément distinguables (en tous cas si on ne dispose que de la taille des sépales).
En particulier il semble possible de séparer les Iris Setosa des autres espèces à partir du paramètre "longueur du pétale" (paramètre n°2)


```python
p_l=X[:,2]
is_setosa=(y==0)
p_l_s=p_l[is_setosa]
max_setosa=p_l_s.max()
p_l_ns=p_l[~is_setosa]
min_non_setosa=p_l_ns.min()
print('Maximum of setosa: {0}.'.format(max_setosa))
print('Minimum of others: {0}.'.format(min_non_setosa))
```

```python
guess_setosa=(p_l<2)
acc= (guess_setosa == is_setosa).mean() # moyenne des coincidences
print ('Performance :{:10.2f} %'.format(acc*100,2))
```

Parfait donc, mais la tâche semblait vraiment aisée. 
Il parait intéressant d'essayer de trouver le "meilleur" seuil pour séparer les Virginica des Versicolor par le même procédé, sachant, d'après la figure précédente, qu'on ne pourra pas réussir à tous coups.

```python
is_virginica=(y==2) # array de booleen
is_versicolor=(y==1)
max_ver=p_l[is_versicolor].max()
min_vir=p_l[is_virginica].min()
best_acc=-1
for t in np.arange(min_vir,max_ver,0.01):
    pred=(p_l>t) #  array de booleen, iris possiblement Virginica
    acc= (pred == is_virginica).mean() # moyenne des coincidences
    if acc > best_acc:
        best_acc=acc
        best_t=t
print('Maximum of Versicolor: {:10.2f}'.format(max_ver))
print('Minimum of Virginica: {:10.2f}'.format(min_vir))        
print ('Meilleur seuil : {:10.2f}'.format(best_t))
print ('Meilleure performance :{:10.2f} %'.format(best_acc*100,2))
```

Il serait bien sûr possible et utile de refaire ce travail sur chacun des 3 autres paramètres, en tous cas, d'après la figure initiale, sur la largeur du pétale , mais il apparait de toute façon impossible d'après cette figure de faire la séparation (sans erreur) entre les deux espèces Versicolor et Virginica en employant cette technique rudimentaire de seuillage sur un paramètre. Voici le programme sous une forme plus aboutie, scindé en deux fonctions.

```python
def learn_thresh_feature(X, y):
    best_acc = -1.0
    # Boucle sur tous les paramètres:
    for fi in range(X.shape[1]):
        max_thresh = X[:, fi].max()
        min_thresh = X[:, fi].min()
        # test sur tous les seuils possibles:
        for t in np.arange(min_thresh,max_thresh,0.01):
            pred = (X[:, fi] > t)
            # Measure the accuracy of this 
            acc = (pred == y).mean()
            if acc > best_acc:
                best_acc = acc
                best_fi = fi
                best_t = t

    # A model is a threshold and an index
    return best_t, best_fi

def accuracy(X, y, t,fi):
    '''Compute the accuracy of the model'''
    preds = X[:, fi] > t
    return np.mean(preds == y)

t,fi=learn_thresh_feature(X,is_virginica)
print ('Meilleur parametre :',feature_names[fi])
print ('Meilleur seuil : {:.2f}'.format(t))
acc=accuracy(X,is_virginica,t,fi)

print ('Meilleure performance : {:.2f} %'.format(best_acc*100,2))
```

### Un peu de rigueur, ne pas mesurer les performances sur la base ayant servi à l'apprentissage (training)

Nous n'avons pas, dans ce qui précède, pris la peine de scinder la base de données en une base d'apprentissage et une base de test. Les performances sont donc sans doute surestimées car le seuil qui permet de faire la distinction entre deux classes a été établi à partir de toutes les données disponibles et pourrait tout à fait varier si on ne considère qu'une partie des données pour l'apprentissage et que l'on mesure la performance (% de bonnes classification) sur le reste des données.


```python
# on enlève les iris setosa de la base car ils sont très faciles à discriminer:
Xv = X[~is_setosa]
yv = y[~is_setosa]
is_vir=(yv==2)
# on divise les données (versicolor et virginica) en deux : testing and training
testing = np.tile([True, False], 50) # testing = [True,False,True,False,True,False...]
training = ~testing

t,fi = learn_thresh_feature(Xv[training], is_vir[training])
train_accuracy = accuracy(Xv[training], is_vir[training], t,fi)
test_accuracy = accuracy(Xv[testing], is_vir[testing], t,fi)

print('''\
Training accuracy : {0:10.2f} %
Testing accuracy : {1:10.2f}%
(N = {2}).
'''.format(train_accuracy*100, test_accuracy*100, testing.sum()))
```

La technique utilisée, le seuillage sur un paramètre, est très rudimentaire. Le prolongement "naturel" de cette technique consisterait à trouver la droite (de direction quelconque) qui séparerait au mieux des données en dimension 2 ou le plan qui séparerait au mieux des données en dimension 3,... C'est ce que réaliseront les techniques de discrimination linéaire.

- 1 Kama

Il s'agit d'une base de dimension assez réduite encore mais qui se prête plus difficilement à une visualisation exhaustive  permettant d'imaginer des traitements triviaux pour séparer les classes (faire la discrimination entre classes).

```python
names=['Area','Perimeter','Compactness','Length of kernel','Width of kernel','Asymmetry Coefficient','Length of kernel groove','Classe']
classes =['Kama','Rosa','Canadian']
import csv
import pandas as pd
res=pd.read_csv("seeds_dataset.txt",delim_whitespace=True,header=None, names=names)
import numpy as np
data=np.array(res)
X=data[:,0:7]
y=data[:,7]# c'est ce dernier paramètre qui est le paramètre de sortie c'est à dire la classe ou label.

plot_croises(X,y-1,classes,names[0:7])
```

### Méthode des plus proches voisins (PPV)

### Méthode des plus proches voisins (PPV)
Nous allons tenter d'utiliser la méthode des plus proches voisins (PPV) pour opérer la discrimination sur cette base. Le principe est simple : pour une donnée dont on ne connait pas la classe mais seulement les caractéristiques, nous allons rechercher la donnée de la base complète (caractéristiques + classe) dont les caractéristiques sont les plus proches (au sens de la distance euclidienne) et nous lui attribuerons la classe de cette donnée la plus proche.

### Séparation de la base en deux sous ensembles formant la based'apprentissage et la base de test

Il est indispensable pour cette technique de plus proches voisins de garder une partie des données pour la validation (la mesure des performances) car si on évalue ces performances sur la base d'apprentissage on obtient un taux d'erreur nul!

#### Mise en oeuvre sous Python

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.cross_validation import train_test_split
knn=KNeighborsClassifier(1)
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X,y) #par défaut 3/4 pour train , 1/4 pour test
```

```python
def perf_knn(Xtrain,ytrain,Xtest,ytest):
    knn.fit(Xtrain, ytrain)
    ctest=knn.predict(Xtest)
    cm = confusion_matrix(ytest, ctest) 
    accuracy=sum(np.diag(cm))/sum(sum(cm))
    erreur=1-accuracy
    print ('matrice de confusion \n',cm)
    print('exactitude = {0:.1%}.'.format(accuracy))
    print('taux d''erreur = {0:.1%}.'.format(erreur))
```

```python
perf_knn(Xtrain,ytrain,Xtest,ytest)
```

Ce taux d'erreur n'est pas très important mais peut être amélioré. En effet les différents paramètres sont probablement inégalement liés à la classe du blé mais leurs poids relatifs dans la mesure de la distance sont liés à leurs différents écart-types.
Examinons les moyennes et écart-types des différents paramètres

```python
np.set_printoptions(formatter={'float': '{: 0.2f}'.format})
print( 'moyenne des parametres : {}'.format(np.mean(X,0)))
print( 'écart type des parametres : {}'.format(np.std(X,0)))
```

Un tout petit ecart (0.01) du troisième paramètre (compactness) modifie grandement ce paramètre (c'est la moitié de l'écart type) alors que ce même écart pour le premier paramètre (Area) est absoluement négligeable. Dans la distance euclidienne considérée ces écarts $(x1-x1')^2$ et  $(x3-x3')^2$ seront pris en compte de façon équivalente.

Etudions la contribution séparée de chaque paramètre à l'erreur de classification.

```python

for fi in range(X.shape[1]):
    
    knn.fit(Xtrain[:,[fi]], ytrain)
    ctest=knn.predict(Xtest[:,[fi]])
    cm = confusion_matrix(ytest, ctest) 
    accuracy=sum(np.diag(cm))/sum(sum(cm))
    erreur=1-accuracy
    print('taux d''erreur sur le paramètre {0} = {1:.1%}.'.format(names[fi],erreur))
    
```

Cette analyse montre d'une part qu'il est nécessaire de prendre en compte plusieurs paramètres pour discriminer de façon satisfaisante les classes et d'autre part que les différents paramètres contribuent de façon très différente à l'efficacité de la discrimination. On observe en particulier que le paramètre Area permet à lui seul une assez bonne discrimination ce qui n'est pas du tout le cas du paramètre Compactness.

### Normalisation des données
Nous proposons alors de normaliser les données. Cette normalisation consiste à soustraire à chaque valeur d'un paramètre sa valeur moyenne et à diviser cette valeur centrée par l'écart type du paramètre.

```python
#Normalisation
Xn = X - np.mean(X,0)
Xn /= np.std(X,0)
```

```python
Xtrain,Xtest,Xntrain, Xntest, ytrain, ytest = train_test_split(X,Xn,y)
print('Performances sur base non normalisée')
perf_knn(Xtrain,ytrain,Xtest,ytest)
print('Performances sur base normalisée')
perf_knn(Xntrain,ytrain,Xntest,ytest)
```

La normalisation a amélioré (le plus souvent, mais ça dépend de la séparation train/test) les performances de cette discrimination par plus proche voisin.
On peut aisément vérifier que cela ne change rien aux contributions séparées (faites le!).

On peut finalement illustrer les propos précédents en représentant les données et leur labels suivant les deux axes sur lesquels ont porté nos remarques (Area et Compactness)

```python

from matplotlib.colors import ListedColormap


def train_plot(X, y):
    N=50
    X_abs = np.linspace(X[:, 0].min(), X[:, 0].max(), N)
    X_ord = np.linspace(X[:, 2].min() , X[:, 2].max(), N)
    xmi,ymi = np.meshgrid(X_abs,X_ord)

    X_test=np.c_[xmi.ravel(), ymi.ravel()]
    knn.fit(X[:,[0,2]],y)
    lab=knn.predict(X_test)
    
    plt.figure(figsize=(10, 10))
    plt.xlabel(names[0])
    plt.ylabel(names[2])
    cmap = ListedColormap([(1., .0, .0), (.0, 1., .0), (.0, .0, 1.)])
    
    plt.scatter(X_test[:,0], X_test[:,1], c=lab, marker='o',alpha=0.5,cmap=cmap)
    
    plt.scatter(X[:, 0], X[:, 2], c=y, alpha=1,cmap=cmap)
    
train_plot(X, y)
train_plot(Xn, y)

```

Dans la première figure, (données non normalisées), le plus proche voisin va être le plus proche suivant la projection sur l'axe des abcisses (même s'il est "loin" suivant la projection sur l'axe des ordonnées), car c'est cet axe des abscisses qui va contribuer le plus à la distance. Dans la seconde figure les deux projections sont prises en compte avec la même importance.
